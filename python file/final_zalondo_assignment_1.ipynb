```python
# making both the column in both the dataset as list so that matching can occur with any error

list1 = offers_training_df_zalondo_moremore['comparison_new'].tolist()
list2 = offers_training_df_aboutyou_moremore['comparison_new'].tolist()
```

```python
# iterating through list1 to extract
# it's closest match from list2

mat1 = []

for i in list1:
    mat1.append(process.extract(i, list2, limit=2))
offers_training_df_zalondo_moremore['matches'] = mat1

offers_training_df_zalondo_moremore.head(10)
```

```python
#adding a temporary file to merge both the dataset to calculate the F1 Score and determine the product matching in 'ellesse' brand

offers_training_df_zalondo_ellesse['tmp'] = 1
offers_training_df_aboutyou_ellesse['tmp'] = 1
df = pd.merge(offers_training_df_zalondo_ellesse, offers_training_df_aboutyou_ellesse, on=['tmp'])
df = df.drop('tmp', axis=1)
df.head(4)
```

```python
# removing the columns which are not required

df1= df[['offer_id_x','comparison_new_x', 'offer_id_y', 'comparison_new_y']]
df1.head()
```

```python
#ellesse brand

from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import f1_score

y_true = df1['comparison_new_x']
y_pred = df1['comparison_new_y']

binarizer = MultiLabelBinarizer()

# This should be your original approach
#binarizer.fit(your actual true output consisting of all labels)

# In this case, I am considering only the given labels.
binarizer.fit(y_true)

print('f1 score for ellesse brand:', f1_score(binarizer.transform(y_true), 
         binarizer.transform(y_pred), 
         average='macro'))
```
